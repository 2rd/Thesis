```python
from imdb import IMDb, IMDbError
from SubDownloader import SubDownloader
import SubDownloader.utils as utl
import pandas as pd
import os
import numpy as np
import pickle
```

```python
# create an instance of the IMDb class
ia = IMDb()

def find_from_imdb_id(imdb_id):
    """ Finds the IMDB Object based on a give IMDB ID """
    try:
        result = ia.get_movie(imdb_id)
    except IMDbError as e:
        print("Something went wrong getting the movie, process aborted.")
        
        raise e

    return result
```

```python
movies = pd.read_csv('../data/ml-20m/movies.csv')
print(movies.shape)
```

```python
movies['imdbId'] = np.nan

```

```python
downloaded_subs = (imdb_id[:-4] for imdb_id in os.listdir('../output/subtitles'))
downloaded_subs = list(downloaded_subs)
# downloaded_subs = list(map(lambda x: float(x), downloaded_subs))
# downloaded_subs = list(map(find_from_imdb_id, downloaded_ids))
# downloaded_subs = list(movie['title'] for movie in downloaded_subs)
```

```python
# match movies in visual feature dataset with movielens dataset
import os

dir = 'D:/Masterprosjekt/Labels'
vis_movies = []

# All movies in the visual feature dataset
for movie in os.listdir(dir):
    vis_movies.append(movie[:-2].lstrip('0'))
vis_movies = list(map(lambda x: float(x), vis_movies))

for row in movies.itertuples():
    curr_id = movies.at[row[0], 'movieId']
    if not curr_id in vis_movies:
        movies = movies.drop(index=row[0])
```

```python
# Add imdb ids to movies

for r in movies.itertuples(index=True):
    if np.isnan(movies.at[r[0], 'imdbId']):
        imdb_search = ia.search_movie(r[2])
        if(len(imdb_search)) != 0:
            id = imdb_search[0].movieID
            movies.at[r[0], 'imdbId'] = id
```

```python
# Remove movies for which subs are already downloaded.

for r in movies.itertuples():
    if r[4] in downloaded_subs or r[2] in downloaded_subs:
        movies.drop(index=r[0], inplace=True)
        
```
